```python
import graphlab as gl
import pandas as pd
import pyLDAvis
import pyLDAvis.graphlab
from collections import Counter
pyLDAvis.enable_notebook()
```

```python
sf = gl.SFrame(data='./parsed_title.csv')
```

```python
docs = [dict(Counter(x)) for x in sf['parsed_title']]
docs = gl.SArray(docs)
# docs
```

```python
# gl.text_analytics.count_words(sf['parsed_title'])
```

```python
topic_model = gl.topic_model.create(docs, beta=0.2, num_topics=20, num_iterations=1000,print_interval=100)
# topic_model.get_topics().print_rows(500)
topic_model
```

```python
lda = pyLDAvis.graphlab.prepare(topic_model,  docs, n_jobs= 6) # mds='tsne',
```

```python
lda
```

```python
pyLDAvis.save_html(lda, 'topic_.html', d3_url='./d3.v3.min.js', ldavis_url='./ldavis.v1.0.0.js', ldavis_css_url='./ldavis.v1.0.0.css')
```

```python
lda_mmds = pyLDAvis.graphlab.prepare(topic_model, docs, mds='mmds' , n_jobs= 4) # mds='tsne',
```

```python
pyLDAvis.save_html(lda_mmds, 'topic_mmds.html',   d3_url='./d3.v3.min.js', ldavis_url='./ldavis.v1.0.0.js', ldavis_css_url='./ldavis.v1.0.0.css')
```
